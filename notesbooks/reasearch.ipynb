```python
rating_with_books=rating.merge(books, on='ISBN')
rating_with_books
```

```python
number_of_rating=rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_of_rating
```

```python
number_of_rating.rename(columns={'rating':'number_of_rating'},inplace=True)
```

```python
number_of_rating.head()
```

```python
final_rating=rating_with_books.merge(number_of_rating, on='title')
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating['number_of_rating']>50]
final_rating.head(3)

```

```python
final_rating.shape
```

```python
final_rating['user_id'].nunique()
```

```python
#book_pivot=final_rating.pivot_table(columns="user_id",index="title",values="rating")
book_pivot = final_rating.pivot_table(columns="user_id", index="title", values="rating")
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[237,:]
```

```python
distance ,suggestion =model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```
